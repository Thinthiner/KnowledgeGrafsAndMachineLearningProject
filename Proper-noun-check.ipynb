```python
## Loop and find words starting with capital letter
temp_list = []
for i in range(0, len(data)):
  if data[i][0].isupper():
    temp_list.append(data[i])
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Chrome("/Users/maikalindkvistjensen/Downloads/chromedriver")
```

```python
(1-(2/13))*100
```

```python
descriptions=[]
words=[]

for search_word in temp_list:
    #driver.get("https://www.wikidata.org/w/index.php?sort=relevance&search={}&title=Special:Search&profile=advanced&fulltext=1&advancedSearch-current=%7B%7D&ns0=1&ns120=1".format(search_word,search_word))
    driver.get("https://www.wikidata.org/w/index.php?title=Special:Search&sort=relevance&search={}&profile=advanced&fulltext=1&advancedSearch-current=%7B%7D&ns0=1&ns120=1".format(search_word,search_word))
    content = driver.page_source
    soup = BeautifulSoup(content)
    if len(soup.findAll(attrs={'class':'wb-itemlink-description'}))>0: #not using if it does not exist in wembedder
        if soup.findAll(attrs={'class':'wb-itemlink-description'})[0].text.lower() == search_word.lower():
            definition = soup.findAll(attrs={'class':'wb-itemlink-description'})[1]
        else:
            definition = soup.findAll(attrs={'class':'wb-itemlink-description'})[0]
        if definition.text != "": #only save if there actually is a description
            words.append(search_word)
            descriptions.append(definition.text)
```

```python
df = pd.DataFrame({'Word':words,'Definition':descriptions}) 
df
```

```python
df.to_csv('ProperNounsDanskNy.csv', index=False, encoding='utf-8')
```
